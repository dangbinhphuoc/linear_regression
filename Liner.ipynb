```
import numpy as np
import matplotlib.pyplot as plt
```

```
'''
Khởi tạo data
'''
x_train = np.arange(1, 50, 2)
y_train = x_train*2 + 1
```

```
'''
Vẽ biểu đồ
'''
plt.plot(x_train, x_train, 'bo')
```

```
'''
Tiền xử lý dữ liệu
'''
x_train = x_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
```

```
'''
Thiết lập model
'''

from sklearn import linear_model
liner = linear_model.LinearRegression()
history = liner.fit(x_train, y_train)
```

```
'''
Dự đoán
y = 2*x + 1
'''
x_test = np.array([10,11,12]).reshape(-1, 1)
y_test = liner.predict(x_test)
print(y_test)
```

```
'''
w
'''
print(history.coef_)
print(history.intercept_)
```

```
'''
Độ chính xác
'''
print(history.score(x_train, y_train))
```
